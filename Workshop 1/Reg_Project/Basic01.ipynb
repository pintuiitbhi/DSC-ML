# STEP 1: Importing the required Libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline 
```


# STEP 2: Loading the Data

```python
df = pd.read_csv('cleanedData.csv') 
```

```python
df.shape #number of rows and columns
```

```python
df.head(5) #Display first five samples in dataframe
```

```python
df = df.drop(['Brand', 'Condition', 'Fuel', 'Model', 'Registered_City', 'Transaction_Type'], axis=1)

#axis=1 : apply methods across columns
```

```python
df.head(5) #Display first five samples in dataframe
```

```python
plt.hist(df['Price'])
plt.show()
```

# STEP 3: Divide the data into training and test data

```python
df_y = df['Price'].values
df_X = df.drop(['Price'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
# from sklearn.modle_selection import train_test_split
```

```python
test_size = 0.30

#Split into train and test
X_train, X_test, Y_train, Y_test = train_test_split(df_X, df_y, test_size = test_size, random_state = 3)
```

# STEP 4: Linear Regression Model

```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error
```

```python
regr = LinearRegression()
```

```python
regr.fit(X_train, Y_train)
```

### Comparison of first five predicted and actual price in train_set

```python
y_pred = regr.predict(X_train)
```

```python
#creating a dataframe
out = pd.DataFrame({'Actual': Y_train, 'Predicted': y_pred})
```

```python
#Basic01
# Plot outputs
out = out.head(5)
out.plot(kind='bar',figsize=(10,6))
plt.show()
```

```python
y_pred = regr.predict(X_test)
```

```python
#creating a dataframe
out = pd.DataFrame({'Actual': Y_test, 'Predicted': y_pred})
```

```python
#Basic01
# Plot outputs
out = out.head(5)
out.plot(kind='bar',figsize=(10,6))
plt.show()
```
