# STEP 3: Data Preprocessing

```python
#Basic03
#Checking missing values in dataset
print(df.isnull().sum(axis = 0))
```

```python
#Basic03
df[df.isnull().any(axis=1)]
```

```python
#Basic03
df=df.dropna() #drop rows with atleast a column with missing values
```

```python
#Basic03
print(df.isnull().sum(axis = 0))
```

```python
#Basic04
#Distribution plot of price
import seaborn as sns
sns.distplot(df["Price"])
df.shape
```

```python
#Basic04
cols=['Price']


for i in cols:
    quartile_1,quartile_3 = np.percentile(df[i],[25,75])
    quartile_f,quartile_l = np.percentile(df[i],[1,99])
    IQR = quartile_3-quartile_1
    lower_bound = quartile_1 - (1.5*IQR)
    upper_bound = quartile_3 + (1.5*IQR)
    print(i,lower_bound,upper_bound,quartile_f,quartile_l)

    df[i].loc[df[i] < lower_bound] = quartile_f
    df[i].loc[df[i] > upper_bound] = quartile_l
```

```python
sns.distplot(df["Price"])
df.shape
```

```python
#Basic04
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out

```

```python
#Basic04
df=remove_outlier(df, 'Price')
df.shape
```

```python
#Basic04
sns.distplot(df["Price"])
```

```python
#Basic02
#One hot encoding
df=pd.get_dummies(df,drop_first=True)
df.head()
```

```python
plt.hist(df['Price'])
plt.show()
```

# STEP 4: Divide the data into training and test data

```python
df_y = df['Price'].values
df_X = df.drop(['Price'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
test_size = 0.30

#Split into train and test
X_train, X_test, Y_train, Y_test = train_test_split(df_X,df_y, test_size=test_size, random_state = 3)
```

# STEP 5: Linear Regression Model

```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error
```

```python
regr = LinearRegression()
```

```python
regr.fit(X_train, Y_train)
```
